```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2

import numpy as np
import pandas as pd

df = pd.read_csv('recordings/02/driving_log.csv', names=('center', 'left', 'right', 'steering', 'throttle', 'break', 'speed'))
```

```python
steering = df['steering']
len(steering)
```

```python
smooth_steering = df.ewm(span=5).mean()['steering']
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
FIGSIZE=(16, 8)
steering.plot.line(figsize=FIGSIZE)
smooth_steering.plot.line(figsize=FIGSIZE)
```

```python
steering.plot.hist(figsize=FIGSIZE, bins=50)
smooth_steering.plot.hist(figsize=FIGSIZE, bins=50)
```

```python
image = mpimg.imread('recordings/02/IMG/center_2017_04_02_14_56_09_576.jpg')
print('This image is:', type(image), 'with dimensions:', image.shape)
plt.imshow(image)
```

```python
def canny(img, low_threshold=50, high_threshold=150):
    kernel_size = 3
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    img = cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)
    return cv2.Canny(img, low_threshold, high_threshold)

plt.imshow(canny(image), cmap='gray')
```
